# 🚀 Convert Fine-Tuned Model to GGUF & Deploy
This notebook will download your fine-tuned model from Hugging Face, convert it into the highly efficient **GGUF** format (Q8_0 quantization), and push it back to your Hugging Face repository so your `llama.cpp` Docker Space can serve it for free!

```
# 1. Install necessary libraries
!pip install huggingface_hub
!git clone https://github.com/ggerganov/llama.cpp.git
!pip install -r llama.cpp/requirements.txt
```

```
# 2. Login to Hugging Face (Paste your write-access token here)
from huggingface_hub import notebook_login
notebook_login()
```

```
# 3. Download your fine-tuned model
# Change this if your repo name is different
REPO_ID = "madhusudhan001/qwen2.5-0.5b-materials-science"
MODEL_DIR = "model_download"

from huggingface_hub import snapshot_download
snapshot_download(repo_id=REPO_ID, local_dir=MODEL_DIR)
```

```
# 4. Convert to GGUF format (Q8_0 quantization is a great balance of size and quality)
!python llama.cpp/convert_hf_to_gguf.py {MODEL_DIR} --outtype q8_0 --outfile model-q8_0.gguf
```

```
# 5. Upload the GGUF file back to your repo
from huggingface_hub import HfApi
api = HfApi()

print("Uploading GGUF model...")
api.upload_file(
    path_or_fileobj="model-q8_0.gguf",
    path_in_repo="model-q8_0.gguf",
    repo_id=REPO_ID,
    repo_type="model"
)
print("✅ Upload complete! Your GGUF model is ready.")
```
